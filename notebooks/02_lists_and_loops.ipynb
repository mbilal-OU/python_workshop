## lists

a list stores multiple values in one variable. like a rack of samples.

```python
samples = ["A1", "A2", "A3", "B1", "B2"]
```

````python
# a list of gene names
genes = ["dnaA", "gyrA", "rpoB", "recA", "lexA"]
print(genes)
print(f"we have {len(genes)} genes")
print(f"first gene: {genes[0]}")    # index starts at 0!
print(f"last gene: {genes[-1]}")   # -1 gives the last one
````

## common mistake - index out of range

lists start at 0. a list of 5 items has indexes 0,1,2,3,4. index 5 doesn't exist.

````python
genes = ["dnaA", "gyrA", "rpoB", "recA", "lexA"]
print(genes[5])
````

````python
# fixed
print(genes[4])   # last item is index 4, not 5
print(genes[-1])  # or just use -1
````

## useful list operations

````python
sequences = ["ATGCGT", "GCTAGC", "TTAAGC"]

# add to list
sequences.append("CCGGTT")
print(sequences)

# sort
sequences.sort()
print(sequences)

# check if something is in the list
print("ATGCGT" in sequences)
print("AAAAAA" in sequences)
````

## for loops

a for loop repeats code for each item in a list. incredibly useful when you have 100 samples and need to do the same thing to each one.

````python
genes = ["dnaA", "gyrA", "rpoB", "recA"]

for gene in genes:
    print(f"processing gene: {gene}")
````

## real biology example - calculating GC content for multiple sequences

````python
sequences = [
    ("seq1", "ATGCGTAAG"),
    ("seq2", "GCGCGCGCG"),
    ("seq3", "ATATAT"),
    ("seq4", "ATGCATGC"),
]

for name, seq in sequences:
    gc = (seq.count("G") + seq.count("C")) / len(seq) * 100
    print(f"{name}: {seq}  GC = {gc:.1f}%")
````

## if / else - making decisions

you'll use this all the time for filtering data

````python
gc_values = [55.6, 100.0, 0.0, 50.0, 66.6, 44.2]

for gc in gc_values:
    if gc > 60:
        print(f"{gc:.1f}% - HIGH GC")
    elif gc < 40:
        print(f"{gc:.1f}% - LOW GC")
    else:
        print(f"{gc:.1f}% - normal range")
````

## common mistake - = instead of ==

this one is classic. in an if statement you need == (comparison), not = (assignment)

````python
gc = 55.6
if gc = 60:
    print("equal")
````

````python
# fixed
gc = 55.6
if gc == 60:
    print("equal to 60")
else:
    print("not equal to 60")
````

## range() - loop a specific number of times

````python
# loop 5 times
for i in range(5):
    print(f"sample number {i+1}")
````
