## basic maths

````python
total_bases = 3280000
gc_bases = 2180000

gc_percent = (gc_bases / total_bases) * 100
print(f"GC content: {gc_percent:.1f}%")

cds_length = 300
protein_length = cds_length // 3
print(f"protein is {protein_length} amino acids long")
````

## string operations - useful for sequences

````python
dna = "atgcgtaagcttgca"

# common string operations
print(dna.upper())           # convert to uppercase
print(len(dna))              # length
print(dna.count("a"))        # count occurrences
print(dna[0:6])              # slicing - first 6 bases
print(dna.replace("t","u"))  # DNA to RNA
````
