````python
import pandas as pd
import matplotlib.pyplot as plt

print(f"pandas version: {pd.__version__}")
````

## creating a DataFrame

a DataFrame is pandas' main data structure. it's basically a table with rows and columns.

````python
# create from a dictionary
data = {
    "organism":    ["D. radiodurans", "E. coli", "T. thermophilus", "B. subtilis", "S. cerevisiae"],
    "domain":      ["Bacteria", "Bacteria", "Bacteria", "Bacteria", "Eukaryote"],
    "genome_mb":   [3.28, 4.60, 2.11, 4.21, 12.07],
    "gc_percent":  [66.6, 50.8, 69.4, 43.5, 38.3],
    "gene_count":  [3195, 4288, 2238, 4176, 6275],
    "habitat":     ["soil", "gut", "hot springs", "soil", "lab strain"],
}

df = pd.DataFrame(data)
print(df)
````

## exploring your data - first things to always do

````python
# shape - how many rows and columns
print(f"shape: {df.shape}  ({df.shape[0]} organisms, {df.shape[1]} columns)")

# column names
print(f"\ncolumns: {list(df.columns)}")

# data types
print(f"\ndtypes:\n{df.dtypes}")

# basic statistics
print(f"\ndescribe:\n{df.describe().round(2)}")
````

## selecting columns and rows

````python
# select one column
print(df["gc_percent"])

# select multiple columns
print(df[["organism", "gc_percent", "genome_mb"]])
````

````python
# filter rows - only bacteria
bacteria = df[df["domain"] == "Bacteria"]
print(bacteria)

# filter rows - GC content above 60%
high_gc = df[df["gc_percent"] > 60]
print("\nhigh GC organisms:")
print(high_gc[["organism", "gc_percent"]])
````

## reading a CSV into pandas

in real work you'll almost always start by reading a file

````python
# read a CSV
df_csv = pd.read_csv("data/survival_data.csv")
print(df_csv)
print(f"\nshape: {df_csv.shape}")
````

## adding new columns - calculated values

````python
# add a survival fraction column
df_csv["od_fraction"] = df_csv["od600"] / df_csv["od600"].max()
print(df_csv[["sample", "treatment", "od600", "od_fraction"]].round(3))
````

## quick statistics

````python
print(f"mean OD600: {df_csv['od600'].mean():.3f}")
print(f"max  OD600: {df_csv['od600'].max():.3f}")
print(f"min  OD600: {df_csv['od600'].min():.3f}")
````

## basic plotting with pandas

````python
# bar chart of GC content
df.plot.bar(x="organism", y="gc_percent", color="steelblue", figsize=(8,4))
plt.title("GC content across organisms")
plt.ylabel("GC content (%)")
plt.xlabel("")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.savefig("data/gc_content_plot.png", dpi=150)
plt.show()
print("plot saved")
````

````python
# scatter plot - genome size vs gene count
df.plot.scatter(x="genome_mb", y="gene_count", figsize=(6,5))

# add organism labels
for i, row in df.iterrows():
    plt.annotate(row["organism"].split()[0], 
                 (row["genome_mb"], row["gene_count"]),
                 textcoords="offset points", xytext=(5,5), fontsize=8)

plt.title("Genome size vs gene count")
plt.xlabel("Genome size (Mb)")
plt.ylabel("Number of genes")
plt.tight_layout()
plt.savefig("data/genome_scatter.png", dpi=150)
plt.show()
````
