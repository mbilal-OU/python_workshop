## functions

a function is a block of code you name and can run whenever you need it.

why use functions?
- avoid repeating the same code over and over
- easier to debug
- easier to read

we've already used functions like `print()`, `len()`, `type()`. now we'll write our own.

```python
# define a function
def calculate_gc(sequence):
    g = sequence.count("G") + sequence.count("g")
    c = sequence.count("C") + sequence.count("c")
    gc_percent = (g + c) / len(sequence) * 100
    return gc_percent

# now use it
seq1 = "ATGCGTAAGCTT"
seq2 = "GCGCGCGCGC"

print(f"seq1 GC: {calculate_gc(seq1):.1f}%")
print(f"seq2 GC: {calculate_gc(seq2):.1f}%")
```

## common mistake - forgetting to return

a function that calculates something but doesn't return the result is one of the most common bugs

```python
def calculate_gc_broken(sequence):
    g = sequence.count("G")
    c = sequence.count("C")
    gc_percent = (g + c) / len(sequence) * 100
    # forgot return!

result = calculate_gc_broken("ATGCGT")
print(result)
```

```python
# fixed - add the return
def calculate_gc_fixed(sequence):
    g = sequence.count("G")
    c = sequence.count("C")
    gc_percent = (g + c) / len(sequence) * 100
    return gc_percent

result = calculate_gc_fixed("ATGCGT")
print(result)
```

## functions with default values

you can give parameters default values so the function works even if you don't pass them

```python
def describe_sequence(sequence, name="unknown"):
    length = len(sequence)
    gc = calculate_gc(sequence)
    print(f"sequence: {name}")
    print(f"  length : {length} bp")
    print(f"  GC     : {gc:.1f}%")
    print()

describe_sequence("ATGCGTAAGCTT", "gene_1")
describe_sequence("GCGCGCGCGC")  # no name given, uses default
```

## dictionaries

a dictionary stores data as **key: value** pairs. like a real dictionary - you look up a word (key) and get its definition (value).

in biology: think of it as a lookup table

```python
# codon table - simplified
codon_table = {
    "ATG": "Met",
    "TAA": "Stop",
    "TAG": "Stop",
    "TGA": "Stop",
    "GCT": "Ala",
    "GCC": "Ala",
    "TGT": "Cys",
    "GAT": "Asp",
}

print(codon_table["ATG"])   # look up ATG
print(codon_table["TAA"])   # look up TAA
```

```python
# add entries
codon_table["AAA"] = "Lys"
codon_table["TTT"] = "Phe"

# check if a codon exists
if "ATG" in codon_table:
    print("ATG is in the table")

# loop through a dictionary
for codon, amino_acid in codon_table.items():
    print(f"{codon} -> {amino_acid}")
```

## real example - storing organism data

```python
organisms = {
    "D. radiodurans": {"genome_mb": 3.28, "gc": 66.6, "habitat": "soil/radiation"},
    "E. coli":        {"genome_mb": 4.60, "gc": 50.8, "habitat": "gut"},
    "T. thermophilus":{"genome_mb": 2.11, "gc": 69.4, "habitat": "hot springs"},
    "B. subtilis":    {"genome_mb": 4.21, "gc": 43.5, "habitat": "soil"},
}

# look up one organism
org = "D. radiodurans"
print(f"{org}:")
print(f"  genome: {organisms[org]['genome_mb']} Mb")
print(f"  GC    : {organisms[org]['gc']}%")
print(f"  habitat: {organisms[org]['habitat']}")
```

```python
# find the organism with highest GC content
highest_gc = 0
highest_org = ""

for org, data in organisms.items():
    if data["gc"] > highest_gc:
        highest_gc = data["gc"]
        highest_org = org

print(f"highest GC: {highest_org} ({highest_gc}%)")
```
