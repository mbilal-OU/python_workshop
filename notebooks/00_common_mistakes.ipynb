# Common Mistakes - Quick Reference

this notebook covers the errors you'll hit most often. if something breaks, look here first.

collected from running this workshop a few times - these come up basically every session.

## 1. NameError - variable doesn't exist

````python
print(my_sequence)
````

````python
# fix: define it first
my_sequence = "ATGCGT"
print(my_sequence)
````

also happens when you run cells out of order in jupyter. if you defined a variable in cell 3 and then restart the kernel, it's gone. run cells from the top.

## 2. SyntaxError - python can't understand your code

````python
# missing colon after if
if len(my_sequence) > 5
    print("long sequence")
````

````python
# fixed
if len(my_sequence) > 5:
    print("long sequence")
````

other common syntax errors:
- `=` instead of `==` in if statement
- missing closing bracket or quote
- wrong indentation

## 3. IndentationError - wrong spacing

````python
for base in "ATGC":
print(base)
````

````python
# fixed - indent code inside the loop
for base in "ATGC":
    print(base)
````

## 4. TypeError - wrong data type

````python
count = 42
message = "found " + count + " genes"
````

````python
# fix 1: convert to string
message = "found " + str(count) + " genes"
print(message)

# fix 2: use f-string (better)
message = f"found {count} genes"
print(message)
````

## 5. IndexError - list index out of range

````python
bases = ["A", "T", "G", "C"]
print(bases[4])  # 4 items, indexes 0-3
````

````python
print(bases[3])   # last item
print(bases[-1])  # also last item
````

## 6. KeyError - key not in dictionary

````python
organism = {"name": "E. coli", "gc": 50.8}
print(organism["habitat"])
````

````python
# fix: use .get() - returns None instead of crashing
print(organism.get("habitat"))
print(organism.get("habitat", "unknown"))  # default value
````

## 8. function returns None

````python
def calculate_length(seq):
    length = len(seq)
    # forgot return!

result = calculate_length("ATGCGT")
print(result)
````

````python
def calculate_length(seq):
    length = len(seq)
    return length  # add return

result = calculate_length("ATGCGT")
print(result)
````

## 9. modifying a list while looping through it

this is a subtle one that causes weird behaviour

````python
# this can skip items or cause errors
numbers = [1, 2, 3, 4, 5]
for n in numbers:
    if n % 2 == 0:
        numbers.remove(n)
print(numbers)  # wrong result - missed some
````

````python
# fix: create a new list instead
numbers = [1, 2, 3, 4, 5]
odd_numbers = [n for n in numbers if n % 2 != 0]
print(odd_numbers)
````

## 10. pandas - SettingWithCopyWarning

this is a common pandas gotcha when filtering a dataframe and then modifying it

if you see: `SettingWithCopyWarning: A value is trying to be set on a copy of a slice`

use `.copy()` when creating a filtered dataframe:

```python
# instead of:
df_filtered = df[df['gc'] > 50]
df_filtered['new_col'] = 'value'  # warning!

# do this:
df_filtered = df[df['gc'] > 50].copy()
df_filtered['new_col'] = 'value'  # no warning
```

## quick error lookup

| error | most likely cause |
|---|---|
| `NameError` | variable not defined, or typo in variable name |
| `SyntaxError` | missing `:`, wrong `=` vs `==`, unclosed bracket |
| `IndentationError` | code not indented inside loop/function/if |
| `TypeError` | mixing strings and numbers without converting |
| `IndexError` | list index >= length of list (remember: starts at 0) |
| `KeyError` | key doesn't exist in dictionary |
| `FileNotFoundError` | wrong file path or file doesn't exist |
| `AttributeError` | calling a method that doesn't exist for that type |
| `ZeroDivisionError` | dividing by zero |
| `ImportError` | library not installed (`pip install libraryname`) |
