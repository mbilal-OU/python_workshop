# Ternary expressions

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)

age = 15
status = "Adult" if age >= 18 else "Minor"
print(status)
```

**Ternary Expressions (Conditional Expressions)**

A ternary expression, also known as a conditional expression, allows you to assign a value to a variable based on a condition in a single line. It's a concise way to write simple `if-else` statements.

**Syntax:**

`value_if_true if condition else value_if_false`

**Explanation of the code above:**

*   `age = 20`: Initializes the `age` variable to 20.
*   `status = "Adult" if age >= 18 else "Minor"`: This is the ternary expression.
    *   `age >= 18` is the condition.
    *   If the condition is `True` (age is 20, which is `>= 18`), then `"Adult"` is assigned to `status`.
    *   If the condition is `False`, then `"Minor"` would be assigned to `status`.
*   The same logic is applied for `age = 15`, resulting in `"Minor"` being assigned to `status`.


When concatenating lists, `list.extend()` is generally more efficient than `+` (the concatenation operator) especially for a large number of concatenations or large lists. This is because `+` creates a new list object each time, which involves reallocating memory and copying elements, while `extend()` modifies the list in place without creating intermediate lists.

```python
list_of_lists = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

# Using extend()
# This is generally more efficient as it modifies the list in place.
everything_extend = []
for chunk in list_of_lists:
  everything_extend.extend(chunk)

print("Using extend():", everything_extend)
```

```python
list_of_lists = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

# Using +
# This creates a new list object in each iteration, which can be less efficient
# for many concatenations.
everything_plus = []
for chunk in list_of_lists:
    everything_plus = everything_plus + chunk

print("Using + operator:", everything_plus)
```
