## 1.1 Variables

a variable is a name you attach to a value. think of it like labelling a tube in the lab.

```python
sample_name = "Sample_A"
concentration = 2.5

```

you pick the name, Python stores the value. simple.

````python
# biology variables
species       = "Deinococcus radiodurans"
genome_size   = 3.28        # Mb
gc_content    = 66.6        # percent
chromosomes   = 2

print(species)
print(f"genome size : {genome_size} Mb")
print(f"GC content  : {gc_content}%")
print(f"chromosomes : {chromosomes}")
````

## 1.8 Functions

a function is a reusable block of code. once you write it, you can use it anywhere.

instead of copy-pasting GC calculation code every time, write it once as a function.

````python
def calculate_gc(sequence):
    """calculate GC content as a percentage"""
    sequence = sequence.upper()
    g = sequence.count("G")
    c = sequence.count("C")
    return (g + c) / len(sequence) * 100

# now use it on anything
print(calculate_gc("ATGCGT"))
print(calculate_gc("gcgcgcgc"))
print(calculate_gc("ATATAT"))
````

the three key parts of a function:
- `def` — tells Python you're defining a function
- `parameters` — the inputs in brackets (here: `sequence`)
- `return` — sends the result back

if you forget `return`, the function gives you `None`:

````python
def gc_broken(sequence):
    g = sequence.count("G")
    c = sequence.count("C")
    result = (g + c) / len(sequence) * 100
    # forgot return!

print(gc_broken("ATGCGT"))   # None :(
````

## 2.2 Reading FASTA Files

the most common file format in molecular biology.
let's write a proper FASTA parser.

````python
# first create a test FASTA file to practice with
fasta_content = """>gene_001 dnaA chromosomal replication initiator
ATGAGTTTTATTATTAGGTGGCGGTACTGGGAATCGCCTTTACTATCGAAGCGATTGCC
GATATTGGCGATGAAGCGATTGAACGTATTGCGCGTATTCGCGATATCGATGCGATTGA
>gene_002 gyrA DNA gyrase subunit A
ATGAGCGATTTAAGCGATTATCGATGGCGATATCGATGCGATTGATGCGATGCGATCGA
TGCGATGCGATTGCGATCGATGCGATTGCGATGCGATTGCGATGCGATCGATGCGATTG
>gene_003 recA recombinase
ATGGCTATTCAGATCACTGATCTGGAGATGGGCGAGATTGATGCGATCGATGCGATCGA
TGCGATGCGATTGCGATGCGATTGCGATGCGATGCGATGCGATTGCGATGCGATTGCG
>gene_004 lexA SOS repressor
ATGAAACGCATTTTCCACATGAACGAAACGAATTTCGCGATTGCGATGCGATCGATGCG
ATTGCGATGCGATCGATGCGATTGCGATGCGATTGCGATGCGATCGATGCGATTGCGATG
"""

with open("data/genes.fasta", "w") as f:
    f.write(fasta_content)

print("FASTA file created: data/genes.fasta")
````

````python
def parse_fasta(filename):
    """
    Read a FASTA file and return a dictionary of {sequence_id: sequence}.
    Handles multi-line sequences automatically.
    """
    sequences = {}
    current_id  = None
    current_seq = []

    with open(filename, "r") as f:
        for line in f:
            line = line.strip()

            if line.startswith(">"):              # header line
                if current_id:                    # save previous sequence
                    sequences[current_id] = "".join(current_seq)
                current_id  = line[1:].split()[0] # take just the ID
                current_seq = []
            elif line:                            # sequence line (skip blank)
                current_seq.append(line.upper())

    if current_id:                                # save the last sequence
        sequences[current_id] = "".join(current_seq)

    return sequences


# read the file
seqs = parse_fasta("data/genes.fasta")

for seq_id, seq in seqs.items():
    print(f"{seq_id:<12} {len(seq):4d} bp   GC = {calculate_gc(seq):.1f}%")
````

## 2.4 Writing Results to a File

always save your results. don't just print them.

````python
# analyse all sequences and save to CSV
seqs = parse_fasta("data/genes.fasta")

results = []
for seq_id, seq in seqs.items():
    results.append({
        "id"        : seq_id,
        "length"    : len(seq),
        "gc_percent": round(calculate_gc(seq), 2),
        "a_count"   : seq.count("A"),
        "t_count"   : seq.count("T"),
        "g_count"   : seq.count("G"),
        "c_count"   : seq.count("C"),
    })

# write CSV
fields = ["id", "length", "gc_percent", "a_count", "t_count", "g_count", "c_count"]
with open("data/sequence_stats.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    writer.writeheader()
    writer.writerows(results)

print("saved to data/sequence_stats.csv")
for r in results:
    print(f"  {r['id']}  {r['length']} bp  GC={r['gc_percent']}%")
````

# Mini Project — Sequence Analyser

let's put everything together. we'll build a script that:

1. reads a FASTA file
2. calculates stats for every sequence
3. filters out low-quality sequences
4. saves a results CSV
5. generates a summary figure

this is a real, useful script. you could use it on actual sequencing output.

````python
# all the functions we've written through the workshop
import pandas as pd
import matplotlib.pyplot as plt
import csv, os

def parse_fasta(filename):
    sequences = {}
    current_id, current_seq = None, []
    with open(filename, "r") as f:
        for line in f:
            line = line.strip()
            if line.startswith(">"):
                if current_id:
                    sequences[current_id] = "".join(current_seq)
                current_id  = line[1:].split()[0]
                current_seq = []
            elif line:
                current_seq.append(line.upper())
    if current_id:
        sequences[current_id] = "".join(current_seq)
    return sequences

def calculate_gc(seq):
    return (seq.count("G") + seq.count("C")) / len(seq) * 100

def analyse_sequence(seq_id, seq):
    return {
        "id"        : seq_id,
        "length"    : len(seq),
        "gc_percent": round(calculate_gc(seq), 2),
        "A"         : seq.count("A"),
        "T"         : seq.count("T"),
        "G"         : seq.count("G"),
        "C"         : seq.count("C"),
        "N"         : seq.count("N"),
        "has_n"     : seq.count("N") > 0,
    }

print("all functions loaded")
````

````python
# step 1 — read sequences
seqs = parse_fasta("data/genes.fasta")
print(f"loaded {len(seqs)} sequences")

# step 2 — analyse all sequences
results = [analyse_sequence(sid, seq) for sid, seq in seqs.items()]
df = pd.DataFrame(results)
print(f"\nbefore filtering:")
print(df[["id", "length", "gc_percent", "has_n"]])
````

````python
# step 3 — filter sequences
# remove: too short (< 50 bp) or has ambiguous bases (N)
df_clean = df[(df["length"] >= 50) & (df["has_n"] == False)].copy()

print(f"after filtering : {len(df_clean)} sequences")
print(f"removed         : {len(df) - len(df_clean)} sequences")
````

````python
# step 4 — save results to CSV
df_clean.drop(columns=["has_n"]).to_csv("data/analysis_results.csv", index=False)
print("results saved to data/analysis_results.csv")

# verify
check = pd.read_csv("data/analysis_results.csv")
print(f"file has {len(check)} rows, {len(check.columns)} columns")
````

````python
# step 5 — summary figures
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# GC content per sequence
axes[0].bar(df_clean["id"], df_clean["gc_percent"],
            color="steelblue", edgecolor="white")
axes[0].axhline(y=50, color="red", linestyle="--", alpha=0.5, label="50%")
axes[0].set_title("GC content per gene")
axes[0].set_ylabel("GC content (%)")
axes[0].set_ylim(0, 80)
axes[0].tick_params(axis="x", rotation=20)
axes[0].legend(fontsize=8)

# base composition stacked bar
bases = df_clean[["id", "A", "T", "G", "C"]].set_index("id")
bases.plot(kind="bar", stacked=True, ax=axes[1],
           color=["#4CAF50","#F44336","#2196F3","#FF9800"],
           edgecolor="white")
axes[1].set_title("Base composition per gene")
axes[1].set_ylabel("Base count")
axes[1].tick_params(axis="x", rotation=20)

plt.suptitle("Sequence Analysis Summary", fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig("data/analysis_summary.png", dpi=150, bbox_inches="tight")
plt.show()
print("figure saved to data/analysis_summary.png")
````

# Common Errors Reference

bookmark this section. if something breaks, look here first.

## quick lookup

| error | most likely cause |
|---|---|
| `NameError` | variable not defined, or typo in variable name |
| `SyntaxError` | missing `:`, wrong `=` vs `==`, unclosed bracket |
| `IndentationError` | code not indented inside loop/function/if |
| `TypeError` | mixing strings and numbers without converting |
| `IndexError` | list index out of range (remember: starts at 0) |
| `KeyError` | key doesn't exist in dictionary |
| `FileNotFoundError` | wrong file path or file doesn't exist |
| `AttributeError` | calling a method that doesn't exist for that type |
| `ZeroDivisionError` | dividing by zero |
| `ImportError` | library not installed |
