# **Dates and Times**



%Y= four digit year

%y=  two digit year

%m=two digit month [01,12]

%d= two digit day

%H=Hour(24 hour) 00,23

%I=Hour(12 hour) 01,12

%M= 2 digit Minute

%S=seconds

%w= week day as integer[0(sunday),6]

%W=week number of year(sunday is considered first day of week and days before first sunday of year are week 0)

%U=week number of year(MONDAY is considered first day of week and days before first MONDAY of year are week 0)

%z=UTC time zone offsewt as HHMM

%F=shortcut for %y-%m_%d (2012-4-18)

%D=shortcut for %m/%d/%Y(04/18/12)


```python
from datetime import datetime,date,time
```

```python
dt=datetime(2011,10,29,20,30,21)
```

```python
dt.day
```

```python
dt.minute
```

```python
#we can get equivalent time by calling function
dt.date()
```

```python
dt.time()
```

```python
#The strftime formats datetime as string
dt.strftime('%m/%d/%y %H:%M')
```

```python
#STRINGS CAN BE CONVERTED(PARSED) INTO DATETIME OBJECTS #make sure capital Y
datetime.strptime("20091031",'%Y%m%d')
```

```python
#REPLACING TIME DATA
dt.replace(minute=0,second=0)
```
