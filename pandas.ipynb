***PANDAS***

A **DataFrame** is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is a widely used data structure in data science and machine learning for storing and manipulating data.

A **dataset** is a collection of related data, typically organized in a structured way. It can be thought of as a single unit of information, often presented in a tabular format (like a spreadsheet or a database table), but can also take other forms such as images, text documents, or audio files. Each row in a tabular dataset usually represents an observation, and each column represents a particular variable or attribute of that observation.
